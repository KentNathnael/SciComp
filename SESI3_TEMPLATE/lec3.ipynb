Basic Plotting (session 3)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
```

```python
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 5, 1, 7, 9, 3, 6, 10]

x = np.array(x)
y = np.array(y)

# y = mx + b
# m => slope
# b => intercept
# y = 2x + 5

plt.plot(x, y, 'bo')
plt.plot(x, 2*x + 5, 'r')
plt.show()
```

```python
# least square refgression

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 5, 1, 7, 9, 3, 6, 10]

x = np.array(x)
y = np.array(y)

a = np.vstack((x, np.ones(len(x)))).T

#y jadi collumn
y = y[:, np.newaxis]

alpha = np.dot(np.dot(np.linalg.inv(np.dot(a.T, a)), a.T), y)

print(alpha)

plt.plot(x,y,'bo')
plt.plot(x, alpha[0]*x + alpha[1], 'r') # y = mx + b
plt.show()
```

```python
# polynomial 

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 5, 1, 7, 9, 3, 6, 10]

x = np.array(x)
y = np.array(y)

for i in range(1,7):
    plt.subplot(2, 3, i)
    plt.plot(x, y, 'bo')

    y_est = np.polyfit(x, y, i)
    print(y_est)
    print(np.polyval(y_est, x))
    print()
    plt.plot(x, np.polyval(y_est, x), 'r')

plt.show()
```

```python
# cubic interpolation

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 5, 1, 7, 9, 3, 6, 10]

x = np.array(x)
y = np.array(y)


f = CubicSpline(x, y)

x_new = np.linspace(1, 8, 100)

# print(x_new)
y_new = f(x_new)
# print(y_new)

plt.plot(x, y, 'bo')
plt.plot(x_new, y_new, 'r')
plt.show()
```
